#### hypothetical scenario where you want to test whether there is a significant difference in the perceived stigma associated with mental health between those who have encountered psychological issues ("Yes") and those who haven't ("No").

```python
from scipy.stats import ttest_ind

# Extract the relevant columns
stigma_encountered = df[df['Encountered Psychological Issue'] == 'Yes']['Stigma']
stigma_not_encountered = df[df['Encountered Psychological Issue'] == 'No']['Stigma']

# Perform independent t-test
t_statistic, p_value = ttest_ind(stigma_encountered, stigma_not_encountered, equal_var=False)

# Print the results
print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")

# Test for statistical significance
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in perceived stigma between those who have encountered psychological issues and those who haven't.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in perceived stigma between those who have encountered psychological issues and those who haven't.")
```

### Test whether individuals with different educational backgrounds have different levels of comfort in discussing mental health issues.

# ANOVA

```python
from scipy.stats import f_oneway

# Extract the relevant columns
education_groups = df['Educational Background'].unique()

# Create a list of comfort scores for each education group
education_data = [df[df['Educational Background'] == edu]['Comfort in disscussion'] for edu in education_groups]

# Perform one-way ANOVA
f_statistic, p_value = f_oneway(*education_data)

# Print the results
print(f"F-Statistic: {f_statistic}")
print(f"P-Value: {p_value}")

# Test for statistical significance
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis.  There is a significant difference in the levels of comfort in discussing mental health issues among individuals with different educational backgrounds.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the levels of comfort in discussing mental health issues among individuals with different educational backgrounds.")
```
